```python
import numpy as np

data = [10, 12, 12, 13, 12, 100, 11, 14, 13, 15]

mean = np.mean(data)
std = np.std(data)

outliers = []
for x in data:
    z = (x - mean) / std
    if abs(z) > 2:
        outliers.append(x)

print("Data:", data)
print("Mean:", mean, "| Std:", std)
print("Outliers detected:", outliers)
```
