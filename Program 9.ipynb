```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

data = {'Value': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

minmax = MinMaxScaler()
standard = StandardScaler()

df['Min-Max'] = minmax.fit_transform(df[['Value']])
df['Z-Score'] = standard.fit_transform(df[['Value']])

print(df)
```
